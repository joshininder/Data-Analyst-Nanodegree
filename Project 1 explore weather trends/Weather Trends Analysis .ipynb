```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Data.csv",index_col='year', parse_dates=True)
df.head()
```

```python
local_moving_average = df.iloc[:,0].rolling(10).mean()
global_moving_average = df.iloc[:,1].rolling(10).mean()
```

```python
plt.plot(local_moving_average, label = "Local Temperature Data")
plt.plot(global_moving_average, label = "Global Temperature")
plt.xlabel("Years")
plt.ylabel("Temperature (in celsius)")
plt.title("Average Temperature for City - Ludhiana")
plt.legend()
plt.show()
```

```python
plt.plot(df["City Average"], label = "Local Temperature Data")
plt.plot(df["Global Average"],label = "Global Temperature")
plt.legend()
```
